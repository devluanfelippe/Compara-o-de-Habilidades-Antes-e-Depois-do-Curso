```python
import matplotlib.pyplot as plt
import seaborn as sns

# Dados do progresso dos alunos
phases = ['Introdução', 'Interface Gráfica', 'Banco de Dados', 'Finalização']
completion_rate = [100, 85, 75, 60]

# Configurando o gráfico
plt.figure(figsize=(8, 6))
sns.barplot(x=phases, y=completion_rate, palette='Blues_d')

# Títulos e rótulos
plt.title('Progresso dos Alunos ao Longo do Curso')
plt.xlabel('Fases do Curso')
plt.ylabel('Percentual de Conclusão (%)')

# Exibir o gráfico
plt.ylim(0, 110)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Dados da avaliação de satisfação
categories = ['Conteúdo', 'Didática', 'Relevância Prática', 'Organização']
ratings = [4.5, 4.2, 4.7, 4.0]

# Configurando o gráfico
plt.figure(figsize=(8, 6))
sns.barplot(x=categories, y=ratings, palette='Greens_d')

# Títulos e rótulos
plt.title('Avaliação de Satisfação dos Participantes')
plt.xlabel('Aspectos Avaliados')
plt.ylabel('Nota (0 a 5)')

# Exibir o gráfico
plt.ylim(0, 5)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Dados de comparação antes e depois
skills = ['Android', 'Java', 'Desenvolvimento de Apps']
before_course = [1.0, 1.5, 0.8]
after_course = [4.0, 4.2, 4.5]

# Configurando o gráfico
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = range(len(skills))

# Barras para antes e depois
plt.bar(index, before_course, bar_width, label='Antes', color='lightcoral')
plt.bar([i + bar_width for i in index], after_course, bar_width, label='Depois', color='lightskyblue')

# Títulos e rótulos
plt.title('Comparação de Habilidades Antes e Depois do Curso')
plt.xlabel('Habilidades')
plt.ylabel('Nível de Conhecimento (0 a 5)')
plt.xticks([i + bar_width / 2 for i in index], skills)

# Exibir a legenda
plt.legend()

# Exibir o gráfico
plt.ylim(0, 5)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```
